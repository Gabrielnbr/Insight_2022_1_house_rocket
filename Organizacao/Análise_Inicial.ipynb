## 0.3.1 Proposição de Mudanças

Observo que não há NA ou NULL no data_set.
1. Modificar as colunas para Português.
2. Modificar a coluna 2 (Data) para tipo data %Y-%M-%D
3. Modificar a coluna "waterfront" para string.
4. Modificar as colunas "sqft" de int para float.
5. Verificar se os ID duplicados fazem parte do fenômeno ou foi algum tipo de erro.
6. Verificar a necessidade dessas duas colunas "sqft_living15" "sqft_lot15", caso desnecessáio para esta análise, descartar.
7. Verificar a necessidade de mudança de pés quadrados "sqft" para m² "m2".
8. Verificar o imóvel com 33 "bedrooms" se realmente há possibilidade de haver 33 quartos ou se pode ser um erro/outlier.
   1. Comparar as casas com um ranged de 10% maior ou 10% menor de m², verificar quantos quartos tem essas casa.

```python
# Modificações realizadas:

data_set['date'] = pd.to_datetime(data_set['date'])

data_set.drop(columns=['sqft_living15','sqft_lot15'], inplace=True)

data_set.loc[data_set['bedrooms'] == 33, 'bedrooms'] = 3

data_set['sqft_living'] = data_set['sqft_living'].astype(float)
data_set['sqft_lot'] = data_set['sqft_lot'].astype(float)
data_set['sqft_above'] = data_set['sqft_above'].astype(float)
data_set['sqft_basement'] = data_set['sqft_basement'].astype(float)
```

```python
data_set['sqft_living'] = data_set['sqft_living'].apply(lambda x: x*0.09290304)
data_set['sqft_lot'] = data_set['sqft_lot'].apply(lambda x: x*0.09290304)
data_set['sqft_basement'] = data_set['sqft_basement'].apply(lambda x: x*0.09290304)
data_set['sqft_above'] = data_set['sqft_above'].apply(lambda x: x*0.09290304)
```

```python
data_set.rename(columns = {'id': 'id', 'date': 'data_venda', 'price':'preco', 'bedrooms':'quartos', 'bathrooms':'banheiros',
                          'sqft_living': 'm2_construido_total','sqft_lot':'m2_terreno_total', 'floors':'andares', 'waterfront':'vista_agua',
                          'view':'vista_geral', 'condition':'condicao', 'grade':'nivel_construcao','sqft_above':'m2_construidos_chao',
                          'sqft_basement':'m2_porao', 'yr_built':'ano_construído', 'yr_renovated':'ano_reformado', 'zipcode':'cep',
                          'lat':'latitude', 'long':'longitude'}, inplace=True)
data_set
```

```python
data_set.loc[data_set['id'].duplicated(keep= False) == True].sort_values(['id','data_venda']).head(20)
```

```python
# Checagem se a data de venda é igual.
# Há uma possibilidade da empresa ter comprado e vendido a mesma casa mais de 1 vez na amostragem. (Atribuir Premissa)
id_duplicado = data_set.loc[(data_set['id'].duplicated(keep= False) == True)].sort_values('id')
print(id_duplicado.loc[id_duplicado['data_venda'].duplicated(keep=False) == id_duplicado['data_venda']].sort_values('id'))
print(id_duplicado.loc[id_duplicado['preco'].duplicated(keep=False) == id_duplicado['preco']].sort_values('id'))
```
